In this activity, I am just using pandas to do some data cleaning

```python
import pandas as pd
```

```python
data = pd.read_csv('dirtydata.csv')
```

```python
data.info()
```

From the information presented above, we can see that Date has 31 rows and calories has 30 rows.
This means that some data are missing in this fields

```python
data.isnull().sum()
```

I used the isnull() and chained with the sum function (dataframe.isnull().sum()) to give me the number of null values there are in the respective columns.

It is apparent from this information above that, Date has 1 null/missing data and Calories has 2

To tackle the Calories column, I chose to calculate the mean of Calories and fill the null values with it.

```python
calories_mean = data['Calories'].mean()
```

```python
data['Calories'].fillna(value=calories_mean, inplace=True)
```

```python
data.isnull().sum()
```

After filling the null values in the Calories column, I checked again to see if the data has been rewritten with the right values.

I am now left with the Date column to deal with. 
Before proceeding, I decided to make sure that all the date fields are of the type of date.
Afterwards, I dropped the null values in the date column by running data.dropna(inplace=True) on the entire data.

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.dropna(inplace=True)
```
